```python
duplicate_rows_df = df[df.duplicated()]
print("number of duplicate rows: ", duplicate_rows_df.shape)
```

```python
df = df.drop_duplicates()
df.head(5)
```

```python
df.count()
```

```python
print(df['Make'].value_counts())
print(df['Vehicle Style'].value_counts())
print(df['Market Category'].value_counts())
print(df['Year'].value_counts())
```

```python
df['Number of Doors'] = df['Number of Doors'].astype('Int64')
df.head()
```

```python
df.isnull().sum()
```

```python
# Null values and percentage of missing data
missing_data = df.isnull().sum()
missing_percent = (missing_data / len(df)) * 100
missing_df = pd.DataFrame({'Missing Values': missing_data, 'Percentage': missing_percent})
missing_df = missing_df[missing_df['Missing Values'] > 0].sort_values(by='Missing Values', ascending=False)
missing_df
```

```python
df.dropna(subset=['Engine HP'], inplace=True)
df.dropna(subset=['Engine Cylinders'], inplace=True)
df.dropna(subset=['Number of Doors'], inplace=True)  
df.dropna(subset=['Engine Fuel Type'], inplace=True)           
```

```python
df.drop(columns=['Market Category'], inplace=True)
df.isnull().sum()
```

```python
plt.figure(figsize=(10,6))
plt.hist(df['MSRP'], bins=30, color='skyblue', edgecolor='black')
plt.title('Distribution of Car Prices (MSRP)')
plt.xlabel('Price ($)')
plt.ylabel('Number of Cars')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.boxplot(x='Vehicle Size', y='MSRP', data=df)
plt.title('Car Price Distribution by Vehicle Size')
plt.xlabel('Vehicle Size')
plt.ylabel('Price ($)')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Group rare fuel types
fuel_counts = df['Engine Fuel Type'].value_counts()
top_fuels = fuel_counts[fuel_counts > 100].index
df['Fuel Type Grouped'] = df['Engine Fuel Type'].apply(lambda x: x if x in top_fuels else 'Other')

# Get counts and labels
fuel_data = df['Fuel Type Grouped'].value_counts().sort_values(ascending=False)
labels = fuel_data.index
counts = fuel_data.values

# Create bar plot
plt.figure(figsize=(10, 6))
bars = plt.bar(labels, counts, color='skyblue', edgecolor='black')

# Add counts on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(
        bar.get_x() + bar.get_width() / 2, 
        height + 5,  # Offset above the bar
        f'{height:,}', 
        ha='center', 
        va='bottom',
        fontsize=10
    )

# Customize plot
plt.title("Engine Fuel Type Distribution", fontsize=14, pad=20)
plt.xlabel("Fuel Type", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.xticks(rotation=45, ha='right')  # Rotate labels for readability
plt.grid(axis='y', linestyle='--', alpha=0.7)  # Add light gridlines

# Adjust layout to prevent clipping
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='Engine HP', y='MSRP', hue='Vehicle Size', data=df)
plt.title('Engine Horsepower vs Car Price')
plt.xlabel('Engine HP')
plt.ylabel('Price ($)')
plt.show()
```

```python
df.Make.value_counts().nlargest(40).plot(kind='bar', figsize=(10,5))
plt.title("Number of cars by make")
plt.ylabel('Number of cars')
plt.xlabel('Make');
```

```python
plt.figure(figsize=(10,8))
correlation = df.corr()
sns.heatmap(correlation, annot=True, cmap='coolwarm', linewidths=0.5)
plt.title('Correlation Matrix')
plt.show()
```
