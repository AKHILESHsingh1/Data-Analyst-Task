##Checking for Missing Values
We use `.isnull().sum()` to identify which columns contain missing data. This helps us decide which columns need cleaning.

```python
print(data.isnull().sum())
```

# Filling missing value
 'director' , 'cast' , 'country' , 'date_added' and 'rating' values with 'Unknown'


```python
data['director'].fillna('Unknown', inplace=True)
```

```python
print(data.isnull().sum())
```

```python
data['cast'].fillna('Unknown', inplace=True)
```

```python
print(data.isnull().sum())
```

```python
data['country'].fillna('Unknown', inplace=True)
```

```python
print(data.isnull().sum())
```

```python
data['date_added'].fillna('Unknown', inplace=True)
```

```python
print(data.isnull().sum())
```

```python
data['rating'].fillna('Unknown', inplace=True)
```

```python
print(data.isnull().sum())
```

# Checking data
changing the datatype of date_added column

```python
data.dtypes
```

```python
data['date_added'] = pd.to_datetime(data['date_added'], errors='coerce')
data.dtypes
```

```python
data.head()
```

```python
print(data.isna().sum())
```

```python
data = data.dropna(subset=['date_added'])
```

```python
data.isna().sum()
```

```python
data.head()
```

# Spliting the listed_in column and creating new column listed_in_split


```python
data['listed_in_split'] = data['listed_in'].str.split(', ')
data.head()
```
